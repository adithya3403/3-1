```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4]
plt.plot(x)
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 4, 6, 8, 10, 12, 14]
z = [1, 6, 9, 12, 15, 18, 21]
print(x, y, z)
plt.plot(x, y, 'ro')
plt.plot(x, z, 'b^')
plt.show()
```

```python
x = np.array([2, 3, 5, 7, 9, 4])
print(x)
y=x**2
print(y)
```

```python
plt.subplot(231)
plt.scatter(x, y)
plt.subplot(232)
plt.bar(x, y)
plt.subplot(233)
plt.plot(x, y)
plt.subplot(234)
plt.hist(x)
plt.subplot(235)
plt.boxplot(x)
plt.subplot(236)
plt.pie(x)
plt.show()
```
