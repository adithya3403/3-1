```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

bias=True for population covaraince

bias=False for sample covaraince

```python
A=[68,60,58,40]
B=[29,26,30,35]
covMatrix=np.cov([A, B], bias=True)
```

# Seaborn

```python
A=[45,37,42,35,39]
B=[38,31,26,28,33]
C=[10,15,17,21,12]
```

```python
data=np.array([A,B,C])
covMatrix=np.cov(data, bias=True)
covMatrix
```

```python
sns.heatmap(covMatrix, annot=True)
plt.show()
```
