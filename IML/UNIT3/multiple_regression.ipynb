```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
df=pd.read_csv('../datasets/co2.csv')
df.head()
```

```python
X=df[['Weight', 'Volume']].values
Y=df['CO2']
```

```python
regr=linear_model.LinearRegression()
regr.fit(X,Y)
```

```python
# predict the CO2 emission of a car where the weight is 2300kg, and the volume is 1300ccm:
predictedCO2=regr.predict([[2300, 1300]])
print(predictedCO2)
```

## Linear Regression - plot the best fit line

```python
slope, intercept, r, p, std_err = stats.linregress(df['Weight'], df['CO2'])

def fun(x):
    return slope * x + intercept

mymodel = list(map(fun, df['Weight']))
plt.scatter(df['Weight'], df['CO2'])
plt.plot(df['Weight'], mymodel)
plt.xlabel("Weight")
plt.ylabel("CO2")
plt.show()
```

```python
slope, intercept, r, p, std_err = stats.linregress(df['Volume'], df['CO2'])

def fun(x):
    return slope * x + intercept

mymodel = list(map(fun, df['Volume']))
plt.scatter(df['Volume'], df['CO2'])
plt.plot(df['Volume'], mymodel)
plt.xlabel("Volume")
plt.ylabel("CO2")
plt.show()
```
